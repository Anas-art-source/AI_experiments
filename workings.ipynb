```python
import pandas as pd
```

```python
df = pd.read_excel('/content/example of a single row with multiple data eletments in a single cell.xlsx')

df
```

```python
data = []
for i, row in df.iterrows():
  first_name = row['First Name']
  last_name = row["Last Name"]
  email = row["Email"]
  list_first_name = first_name.split('\n')
  list_last_name = last_name.split('\n')
  list_email = email.split('\n')
  for i in range(len(list_first_name)):
    data.append({
        "first_name": list_first_name[i],
        'last_name': list_last_name[i],
        'email': list_email[i]
    })
```

```python
df = pd.DataFrame(data)
df.to_csv('upwork_proposal_algo_dump.csv')
```
